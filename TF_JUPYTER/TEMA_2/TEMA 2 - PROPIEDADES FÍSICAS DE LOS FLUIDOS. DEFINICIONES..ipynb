<h1 style="text-align: center;"><div id="definicion"><strong><span style="font-size: 20px;">DEFINICIÓN DE FLUIDO. SÓLIDOS, LÍQUIDOS Y GASES. ANALOGÍAS Y DIFERENCIAS.</span></strong></div></h1>
<hr>
<p style="text-align: justify;">El material s&oacute;lo puede presentarse en dos estados, s&oacute;lido o fluido. La distinci&oacute;n se hace en funci&oacute;n de la reacci&oacute;n de las sustancias frente al esfuerzo cortante.</p>
<br>
<p style="text-align: justify;">Un s&oacute;lido puede resistir un esfuerzo cortante con una deformaci&oacute;n est&aacute;tica; un fluido no. Cualquier esfuerzo cortante aplicado a un fluido provocar&aacute; el movimiento del fluido. Este se mueve y se deforma continuamente mientras se siga aplicando el esfuerzo cortante.</p>
<br>
<img src="MEDIA/2.1.png" style="height:200px">
<br>
<p style="text-align: justify;">Si se aplica una fuerza cortante sobre un s&oacute;lido y sobre un fluido, en ambos casos el esfuerzo cortante ser&aacute; $\tau=F/A$, siendo A la superficie de contacto. El s&oacute;lido se deformar&aacute; un &aacute;ngulo ($\alpha$) y soportar&aacute; el esfuerzo al que se ve sometido. En el fluido, el esfuerzo causar&aacute; que la placa superior se mueva con velocidad uniforme ($v\neq 0$), sin que importe lo peque&ntilde;a que sea. Mientras se ejerza la fuerza, la deformaci&oacute;n continuar&aacute;.</p>
<br>
<p style="text-align: justify;">El s&oacute;lido adquiere una deformaci&oacute;n est&aacute;tica, resistiendo esfuerzos cortantes sin fluir. Contrariamente, el fluido en reposo necesita paredes para eliminar el esfuerzo cortante. Si se quitan las paredes, se crea un esfuerzo cortante y el fluido se derramar&aacute; o se expandir&aacute;.</p>
<br>
<img src="MEDIA/2.2.png" style="height:200px">
<br>
<p style="text-align: justify;">Se ha conseguido distinguir entre s&oacute;lido y fluido, sin embargo existen dos tipos de fluidos, los l&iacute;quidos y los gases; el par&aacute;metro que los diferencia es la distancia intermolecular:</p>
<ul>
<li>Un l&iacute;quido esta compuesto por agrupaciones de mol&eacute;culas muy cercanas, lo que hace que se produzcan enormes fuerzas de cohesi&oacute;n, con lo que tiende a conservar su volumen y, en el caso en que se encuentre en un recipiente no limitado por la parte superior, en un campo gravitatorio formar&aacute; una superficie libre.<br /><strong>Dada la cercan&iacute;a entre mol&eacute;culas, los l&iacute;quidos resultan pr&aacute;cticamente incompresibles.</strong></li>
<li>En el caso del gas las mol&eacute;culas se encuentran muy separadas, con fuerzas cohesivas despreciables, por lo que un gas es libre de expandirse hasta que encuentre las paredes que lo confinan.<br/><strong>Dada esta dispersi&oacute;n molecular los gases resultan compresibles.</strong></li>
</ul>
<br>

<h1 style="text-align: center;"><div id="medio"><strong><span style="font-size: 20px;">EL FLUIDO COMO MEDIO CONTINUO</span></strong></div></h1>
<hr>
<p style="text-align: justify;">Al tratar el flujo de los fluidos matem&aacute;ticamente, se necesita considerar que la estructura molecular original se remplaza por un medio hipot&eacute;tico llamado medio continuo, lo cual significa que puede emplearse el c&aacute;lculo diferencial en su an&aacute;lisis.</p>
<p style="text-align: justify;"><br />Si se considera, la velocidad en un punto del fluido, ser&aacute; cero todo el tiempo excepto cuando una mol&eacute;cula ocupa ese punto; entonces ser&aacute; la velocidad de la mol&eacute;cula. Este dilema se evita si se considera la velocidad en un punto como el promedio de la velocidad de masa de todas las mol&eacute;culas que rodean el punto, por ejemplo, dentro de una peque&ntilde;a esfera con radio grande comparado con la distancia media entre mol&eacute;culas. 
<br>
<br>
La densidad, volumen espec&iacute;fico, presi&oacute;n velocidad y aceleraci&oacute;n se supone que var&iacute;an continuamente en todo el fluido (o que son constantes).</p>

<h1 style="text-align: center;"><div id="viscosidad"><strong><span style="font-size: 20px;">VISCOSIDAD</span></strong></div></h1>
<hr>
<p style="text-align: justify;"><strong>Definicion:</strong> La viscosidad se puede definir como la propiedad de un fluido por la cual &eacute;ste ofrece resistencia al cambio de forma por la acci&oacute;n de fuerzas exteriores &oacute; la propiedad del fluido por la cual &eacute;ste ofrece resistencia a los esfuerzos cortantes.</p>
<img src="MEDIA/2.4.png" style="height:200px">
<p style="text-align: justify;">Supongamos un l&iacute;quido viscoso (ej. aceite) confinado entre dos placas planas paralelas,la inferior fija y la superior m&oacute;vil. Si se ejerce una fuerza tangencial F~ en la placa superior, aparecer&aacute; un esfuerzo cortante &tau; = F/A. Dada la definici&oacute;n de fluido, la placa superior comenzar&aacute; a moverse con una velocidad v uniforme (v distinta de cero).</p>
<p style="text-align: justify;">Un hecho experimental ampliamente demostrado es que la velocidad de las part&iacute;culas fluidas en contacto directo con superficies s&oacute;lidas es la velocidad de dichos contornos s&oacute;lidos.</p>
<p style="text-align: justify;"><br />Entonces, tendremos una velocidad v igual a la de la placa superior en los puntos en contacto con dicha placa y una velocidad nula en los puntos en contacto con la placa inferior fija, es decir, aparecer&aacute; una variaci&oacute;n de la velocidad del fluido de modo que aumenta conforme nos alejemos de la placa fija, como consecuencia del esfuerzo al que se somete al fluido. Es decir, aparecer&aacute; un gradiente de velocidad que ser&aacute; proporcional al esfuerzo al que sometemos al fluido.</p>
<p>&nbsp;</p>
<p style="text-align: justify;">Seg&uacute;n postula la ley de Newton de la viscosidad, ese factor de proporci&oacute;n es precisamente la viscosidad din&aacute;mica, (&mu;). Como consecuencia la expresi&oacute;n de la ley de Newton de la viscosidad ser&aacute;:</p>
<p>&nbsp;</p>
    
<p style="text-align: center;">$${\tau}={\mu}{\cdot}\textit{grad}\ v=\frac{F}{A}$$</p>
<p style="text-align: justify;">En el caso del ejemplo la única variación de velocidad se produce en dirección perpendicular a la placa fija. Si definimos la variable y en dicha dirección, tendremos la siguiente expresión de la ley de Newton de la viscosidad:</p>
<p style="text-align: center;">$${\tau}=\frac{dF}{dA}=\mu\frac{dv}{dy}$$</p>
<p>&nbsp;</p>
<p style="text-align: justify;">A los fluidos que cumplen la ley de Newton de la viscosidad se les denomina fluidos newtonianos.</p>
<br>
<p style="text-align: left;"><span style="text-decoration: underline;"><strong>UNIDADES DE LA VISCOSIDAD DINAMICA</strong></span></p>
<br />
Sistema Internacional: ${\frac{N\cdot m}{m^{{2}}\cdot {m}{/}{s}}=N\cdot m^{{-2}}\cdot
s={\text{kg}}\cdot {\frac{m}{s^{{2}}}}\cdot
m^{{-2}}\cdot s={\text{kg}}\cdot m^{{-1}}\cdot s^{-1}}$ =1 Poiseuille=1 Pl= Pa${\cdot}$s
<br>
<br />Sistema Cegesimal:  $Dyn{\cdot}cm^{-2}{\cdot}s $ = 1
Poise=1 Po= 0.1 Pl
<br>
<br>
<p style="text-align: left;"><span style="text-decoration: underline;"><strong>VARIACION DE LA VISCOSIDAD CON LA TEMPERATURA</strong></span></p>
<br>
<p style="text-align: justify;">En los <strong>líquidos</strong> newtonianos la viscosidad disminuye al aumentar la temperaturade acuerdo, de manera aproximada, con la ecuaciónde Arrhenius:
<br>
<p style="text-align: center;">${\mu} = A\cdot e^{B/T}$ </p>
<br>
Donde T es la temperatura absoluta y A y B son constantes del líquido. En general,
para los líquidos newtonianos, cuanto mayor sea la viscosidad, mayor es la dependencia
de ésta con la temperatura.
<br />
Debido a la cercan&iacute;a entre sus mol&eacute;culas, la cohesi&oacute;n molecular es la causa preponderante de la viscosidad.
<br>
<p style="text-align: justify;">La viscosidad en los <strong>Gases</strong>, a presión constante, varía con la temperatura de acuerdo
con la ley de Sutherland:</p>
<p style="text-align: center;">
\begin{equation*}
{\mu _{{T}}=\mu
_{{0}}\cdot \sqrt{\frac{T}{T_{{0}}}}\cdot \frac{1+C/{T_{{0}}}}{1+C/{T}}}
\end{equation*}</p>
donde T es la temperatura absoluta en grados Kelvin y C una constante
característica de cada gas. ${\mu}_{0}$ es la viscosidad
del gas a 0ºC y ${\mu}_{T}$ la
correspondiente a la temperatura T.
<br />En los gases, la causa preponderante de la viscosidad es la transferencia de cantidades de movimiento moleculares.</p>
<p style="text-align: justify;">Variaci&oacute;n de la viscosidad con la presi&oacute;n: En los l&iacute;quidos, la viscosidad din&aacute;mica aumenta exponencialmente con la presi&oacute;n. Los cambios son bastante peque&ntilde;os para presiones que difieran de la presi&oacute;n atmosf&eacute;rica en alrededor de 1 bar.</p>
<p style="text-align: justify;">Por tanto, el efecto de la presi&oacute;n es ignorado. Hay situaciones, sin embargo, en las que no se puede ignorar el efecto de la presi&oacute;n. Por ejemplo, la industria del petr&oacute;leo requiere medidas de la viscosidad de los lubricantes y de los fluidos de perforaci&oacute;n a presiones elevadas.<br />En los gases, para una temperatura dada, la viscosidad es independiente de la densidad y, por tanto, de la presi&oacute;n.</p>
<br>
<br>
<p style="text-align: justify;">
<span style="text-decoration: underline;"><strong>VARIACION DE LA VISCOSIDAD CON LA PRESION</strong></span></p>
<br>
En los líquidos, la viscosidad dinámica
aumenta exponencialmente con la presión. Los cambios son bastante pequeños
para presiones que difieran de la presión atmosférica en alrededor de 1 bar.
Por tanto, el efecto de la presión es ignorado.
Hay situaciones, sin embargo, en las que no se puede ignorar el efecto de la
presión. Por ejemplo, la industria del petróleo requiere medidas de la viscosidad
de los lubricantes y de los fluidos de perforación a presiones elevadas.
En los gases, para una temperatura dada, la viscosidad es independiente de la
densidad y, por tanto, de la presión.
<br>
<br>
<br>
<span style="text-decoration: underline;"><strong>VISCOSIDAD CINEMATICA</strong></span></p>
<p style="text-align: justify;">Una medida de viscosidad muy utilizada cuando existen flujos es la viscosidad cinem&aacute;tica que es la raz&oacute;n entre la viscosidad din&aacute;mica y la densidad del fluido:</p>
<p style="text-align: justify;">$$ \nu=\frac{\mu}{\rho}$$</p>
<p style="text-align: justify;">Unidades:<br />Sistema Internacional: $1 m^2/s$<br />Sistema Cegesimal: $1 cm^2/s = 1$ Stoke (St)</p>
<p style="text-align: justify;">&nbsp;</p>
<span style="text-decoration: underline;"><strong>VISCOSIDADES EMPIRICAS</strong></span></p>
<p style="text-align: justify;">En muchas aplicaciones industriales es suficiente determinar la viscosidad relativa, tambi&eacute;n llamada viscosidad emp&iacute;rica, que es la viscosidad del l&iacute;quido objeto de ensayo medida comparativamente a la de un l&iacute;quido patr&oacute;n, que frecuentemente es el agua, realiz&aacute;ndose la comparaci&oacute;n en condiciones perfectamente definidas de temperatura constante. Algunos de los viscos&iacute;metros m&aacute;s utilizados son:</p>
<p style="text-align: justify;"><ul>
<li><strong> Engler</strong>: Viscosidad a T °C de un líquido a la relación
de los tiempos de vaciado (en un viscosímetro Engler) de 200 cm3 del líquido a la
temperatura T y del mismo volumen de agua destilada a 20 °C</li>
<br>
<li><strong> Saybolt</strong>: El viscosímetro Saybolt (U.S.A.) mide directamente en segundos el tiempo
de vaciado de 60 cm3 del líquido a ensayar</li>
<br>
<li><strong> Redwood</strong>: El viscosímetro Redwood (U.K.) mide directamente en segundos el
tiempo de vaciado de 50 cm3 del líquido a ensayar.</li>
</ul>

<h1 style="text-align: center; "><div id="modulo"><strong><span style="font-size: 20px;">ELASTICIDAD Y M&Oacute;DULO DE ELASTICIDAD VOLUM&Eacute;TRICO</span></strong></div></h1>
<p style="text-align: justify;">En la mayor&iacute;a de los casos, un l&iacute;quido se puede considerar incompresible,
pero para situaciones en las que se producen cambios repentinos o grandes de presi&oacute;n, su compresibilidad es importante.
Si llamamos $V$ al volumen inicial de una masa fluida sobre cuya superficie l&iacute;mite existe una presi&oacute;n $P$ y 
dicha presi&oacute;n experimenta un incremento $dp$, se producir&aacute; una disminuci&oacute;n de volumen $-dV$. 
La deformaci&oacute;n volum&eacute;trica unitaria es proporcional al incremento de presi&oacute;n $dp$, siendo $\alpha$ el coeficiente de proporcionalidad al que se denomina coeficiente de compresibilidad c&uacute;bica.</p>
<p style="text-align: center;">$$ \frac{dV}{V}=-\alpha·dp $$ </p>
<p style="text-align: justify;">Normalmente se trabaja con el inverso del coeficiente de compresibilidad c&uacute;bica, denominado M&oacute;dulo de Elasticidad Volum&eacute;trico, K.</p>
<p style="text-align: center;">$$ K=\frac{1}{\alpha}=\frac{-dp}{\frac{dV}{V}} $$ </p>

<h1 style="text-align: center;"><div id="capilaridad"><strong><span style="font-size: 20px;">TENSIÓN SUPERFICIAL. CAPILARIDAD.
    </span></strong></div></h1>
<p style="text-align: justify;">Un líquido, al no ser capaz de expansionarse libremente, formará una interfase con
un segundo líquido (inmiscible) o un gas.
<br><br>
Si se analiza la superficie libre de un líquido ( contacto líquido/aire), en el interior del
líquido una molécula es atraída por igual en todas las direcciones por las moléculas que le
rodean, pero en la superficie libre, las atracciones entre las moléculas de aire y del líquido
no están equilibradas. La superficie del líquido se comporta como si fuera una membrana
elástica bajo tensión. La tensión superficial es la fuerza necesaria para mantener la unidad
de longitud de membrana en equilibrio.</p>
<p style="text-align: center;"> $$ dW=\sigma·dA \rightarrow \sigma=\frac{dW}{dA}=\frac{F·dx}{l·dx} \rightarrow \sigma=\frac{F}{l}$$</p>
<p style="text-align: justify;">Unidades:<br>  
    Sistema Internacional: $N/m$<br>  
    Sistema cegesimal: $dyn/cm$<br>  
    Sistema Tecnico: $Kp/m$<br>
<img src="MEDIA/2.5.png" style="height:200px">
<p style="text-align: justify;">La tensión superficial hace que los líquidos tiendan a tomar formas esféricas, resulta
un fenómeno importante en la formación de gotas, pulverización y en el fenómeno de
la capilaridad que se estudia a continuación. Un ejemplo en el que debe considerarse
la tensión superficial es la mezcla gasolina/aire en el carburador de un motor de explosión</p>
    
<span style="text-decoration: underline;"><strong>CAPILARIDAD</strong></span>
<br>
<p style="text-align: justify;">La atracción capilar es causada por la tensión superficial y por el valor relativo de
la adhesión entre líquido y sólido a la cohesión del líquido. Por la acción de la tensión
superficial el líquido asciende (o desciende) dentro de un tubo vertical que se sumerge
parcialmente en él. La comparación de las fuerzas de cohesión del líquido con las de
adhesión del líquido al sólido produce dos situaciones diferentes:</p>

<ul><li><strong>Fuerza adhesión líquido/sólido > Fuerza cohesión líquido</strong><br><br>
    Se dice que el líquido moja al sólido, el nivel interno es superior al externo, y la
    superficie interna es cóncava, se forma el menisco de elevación.<br><br>
    Ejemplo: agua y vidrio.<br><br>
    <li><strong>Fuerza adhesión líquido/sólido < Fuerza cohesión líquido </strong><br><br>
    El líquido no moja, la superficie libre es convexa y el menisco formado es de
    depresión.<br><br>
    Ejemplo: Mercurio y vidrio.

<h1 style="text-align: center;"><div id="cavitacion"><strong><span style="font-size: 20px;">TENSIÓN DE VAPOR. CAVITACIÓN.
</span></strong></div></h1>
<p style="text-align: justify;">Los líquidos se evaporan debido a que las
moléculas se escapan de la superficie del líquido.
Las moléculas de vapor ejercen una presión
parcial en el espacio conocida como presión de
vapor. Si el espacio por encima de la superficie
libre del líquido es limitado (recipiente cerrado),
después de un tiempo suficiente el número de
moléculas de vapor que golpean la superficie del
líquido y se condensan es justamente igual al
número que se escapa en cualquier intervalo de
tiempo, dándose así el equilibrio. Ya que este
fenómeno depende de la actividad molecular,
que es función de la temperatura, la presión de
vapor de un fluido dado depende de la
temperatura y aumenta con ella.
<br><br>
    Cuando la presión por encima de un líquido es
igual a la presión de vapor del líquido ocurre la
ebullición.
<br><br>    
En muchas situaciones, es posible que se produzcan presiones muy
bajas en ciertos puntos del sistema. En tales circunstancias las presiones
pueden ser iguales a la presión de vapor, o menores y el líquido se
evapora rápidamente.
<br><br>
Es el fenómeno de la <strong>cavitación</strong>. Se forma una bolsa o cavidad de vapor
en rápida expansión, que generalmente es barrida lejos de su punto de
origen y penetra regiones donde la presión es mayor que la presión de
vapor, la cavidad se desploma. Este fenómeno afecta a conducciones y
máquinas hidráulicas.
<img src="MEDIA/2.11.png" style="width:80%">
    <p style="text-align: center;">Cavitación en bomba centrífuga
