<h1 style="text-align: center;"><div id="objeto"><strong><span style="font-size: 20px;">OBJETO Y APLICACIONES DE LA MECANICA DE FLUIDOS</span></strong></div></h1>
<br>
<hr>

    
<p style="text-align: justify;">La Mec&aacute;nica de Fluidos se ocupa del estudio del comportamiento de los fluidos, tanto en reposo como en movimiento y de los efectos que &eacute;stos puedan producir sobre su entorno, que puede ser s&oacute;lido u otro fluido.</p>
<p style="text-align: justify;">Dos impedimentos, geometr&iacute;a y viscosidad, que hacen inviable (en la mayor&iacute;a de casos pr&aacute;cticos) su tratamiento puramente te&oacute;rico.</p>
<ul>
<li>Se utilizan geometr&iacute;as sencillas como son las placas planas y los conductos circulares.</li>
<li>Resulta necesario acudir a la experimentaci&oacute;n.</li>
</ul>
<p>&nbsp;</p>
<p style="text-align: justify;">La tierra est&aacute; cubierta en un 75% por agua y en un 100% por aire, lo que hace que las posibilidades de la Mec&aacute;nica de Fluidos sean enormes.</p>
<ul>
<li>Flujos naturales, meteorolog&iacute;a, oceanograf&iacute;a y la hidrolog&iacute;a,...</li>
<li>Flujos biol&oacute;gicos, ej. flujo sangu&iacute;neo y la respiraci&oacute;n,...</li>
<li>Flujos artificiales, construidos por el hombre:
<ul>
<li>Transporte de pasajeros &#8594 hidrodin&aacute;mica (barcos, submarinos) y aerodin&aacute;mica (aviones, cohetes),</li>
<li>Producci&oacute;n de energ&iacute;a &#8594 turbinas hidr&aacute;ulicas o de vapor.</li>
<li>Transporte de fluidos &#8594 regad&iacute;os, abastecimiento de agua, transporte de petr&oacute;leo o gas.</li>
<li>Aplicaciones de la hidrost&aacute;tica &#8594 construcci&oacute;n de presas y dep&oacute;sitos para el almacenamiento de agua, diques de contenci&oacute;n, etc.</li>
</ul>
</li>
</ul>

<h1 style="text-align: center;"><div id="dimensiones"><strong><span style="font-size: 20px;">SISTEMA DE UNIDADES.DIMENSIONES</span></strong></div></h1>
<hr>
<p style="text-align: justify;">Se define la dimensi&oacute;n como la medida por la cual una variable f&iacute;sica se expresa cuantitativamente y la unidad ser&aacute; la forma de asignar un n&uacute;mero a dicho valor.</p>
<p style="text-align: justify;">Las dimensiones son esenciales para poder cuantificar (medir) los diferentes fen&oacute;menos f&iacute;sicos, pero, en la Mec&aacute;nica de Fluidos, adem&aacute;s son la base de una importante herramienta denominada An&aacute;lisis Dimensional cuya aplicaci&oacute;n es cada vez m&aacute;s extendida.</p>
<p style="text-align: justify;">Se pueden dividir en dos tipos: dimensiones primarias (fundamentales) y dimensiones secundarias. Las dimensiones secundarias son aquellas que pueden expresarse en funci&oacute;n de las primarias. Las dimensiones primarias en la Mec&aacute;nica de Fluidos son masa (M), longitud (L), tiempo (T) y temperatura (&theta;).</p>
<p style='background:#DDD9C3;'>Nota: En el Sistema T&eacute;cnico en lugar de la masa se emplea la fuerza como dimensi&oacute;n fundamental y la unidad correspondiente ser&aacute; el kilopondio (kp).</p>

<h1 style="text-align: center;"><div id="densidad"><strong><span style="font-size: 20px;">DENSIDAD, PESO ESPECIFICO Y VOLUMEN ESPECIFICO</span></strong></div></h1>
<hr>   
<p style="text-align: justify;">La densidad <b>&rho;</b> de un fluido se define como su masa por unidad de volumen:</p>
<p style="text-align: center;">$\rho=\frac{dm}{dV}$ &emsp; Unidad (SI): $kg/m^3$</p>
<p style="text-align: justify;">Si el fluido puede considerarse homog&eacute;neo:</p>
<p style="text-align: center;">$\rho=\frac{m}{V}$ </p><br>
<p style="text-align: justify;">La densidad de un fluido var&iacute;a con la presi&oacute;n y con la temperatura. Por ejemplo para el agua en condiciones est&aacute;ndar (p=760 mmHg, T=4&ordm;C) la densidad toma el valor de 1000 kg/m3.</p>
<p style="text-align: justify;">El volumen espec&iacute;fico <b>$V_s$</b> es el inverso de la densidad, es el volumen ocupado por la unidad de masa de fluido. El peso espec&iacute;fico g de una sustancia es su peso por unidad de volumen. Debe tenerse en cuenta que al tratarse de un peso depender&aacute; de la aceleraci&oacute;n de la gravedad.</p>
<p style="text-align: center;">$V_s=\frac{1}{\rho}=\frac{V}{m}$ &emsp; Unidad (SI): $m^3/kg$</p>
<p style="text-align: justify;">El peso espec&iacute;fico <b>&gamma;</b> de una sustancia es su peso por unidad de volumen. Debe tenerse en cuenta que al tratarse de un peso depender&aacute; de la aceleraci&oacute;n de la gravedad.</p>
<p style="text-align: center;">$\gamma=\frac{m\cdot g}{V}=\rho \cdot g$ &emsp; Unidad (SI): $N/m^3$</p>
<p style="text-align: justify;">Otra forma de trabajar con densidades es la de emplear densidades relativas <b>s</b>, que se definen como la raz&oacute;n de su peso al peso de un volumen igual de un l&iacute;quido patr&oacute;n en condiciones est&aacute;ndar. Tambi&eacute;n puede expresarse como una raz&oacute;n de su densidad o su peso espec&iacute;fico al del l&iacute;quido patr&oacute;n. Generalmente el l&iacute;quido patr&oacute;n suele ser el agua.</p>
<p style="text-align: center;">$$s=\frac{W_{sustancia}}{W_{patron}}=\frac{{\rho}_{sustancia}}{{\rho}_{patron}}=\frac{{\gamma}_{sustancia}}{{\gamma}_{patron}}$$</p>

<h1 style="text-align: center;"><div id="termo"><strong><span style="font-size: 20px;">VARIABLES TERMODINAMICAS. ECUACIONES DE ESTADO.</span></strong></div></h1>
<hr>     
<p style="text-align: justify;">Las propiedades termodin&aacute;micas m&aacute;s importantes desde el punto de vista de la Mec&aacute;nica de Fluidos son la presi&oacute;n p, la densidad d, la temperatura T y la viscosidad w.Adem&aacute;s influyen junto con estas la energ&iacute;a interna, la entalp&iacute;a, la entrop&iacute;a, los calores espec&iacute;ficos y la conductividad t&eacute;rmica.</p>
<p style="text-align: justify;">Si el fluido de trabajo es un gas, puede emplearse la Ecuaci&oacute;n de Estado para gases. Todos los gases a altas temperaturas y bajas presiones (relativas a su punto cr&iacute;tico) cumplen la ley de los gases perfectos.</p>
<p style="text-align: center;">$$ P \cdot V=n\cdot R\cdot T $$</p>
<p style="text-align: justify;"><br/>Donde R es la constante universal de los gases perfectos y cuyo valor en unidades del sistema internacional es: R=8314 J/(kmol&middot;K) y n es el n&uacute;mero de moles con los que se est&aacute; trabajando.</p>
<p style="text-align: justify;">Se puede obtener una forma m&aacute;s &uacute;til para la Mec&aacute;nica de Fluidos de la ley de los gases perfectos sustituyendo n en funci&oacute;n de las masas que en ella intervienen:</p>
<p style="text-align: center;">$$ P=\rho \cdot R^{*} \cdot T $$</p>
<p style="text-align: justify;">Donde R* ser&aacute; la relaci&oacute;n entre la constante de los gases R y la masa molecular del gas con el que se est&aacute; tratando M, que para el caso del aire, por ejemplo, es de 287 J/kg&middot;K.</p>

<h1 style="text-align: center;"><div id="gradiente"><strong><span style="font-size: 20px;">CONCEPTO DE GRADIENTE</span></strong></div></h1>
    
<p style="text-align: justify;">Desde el punto de vista de la Fluidomecánica no son más que herramientas matemáticas; por ello este apartado se limita a su
presentación.<br><br>
Se representa mediante <strong>grad</strong> o ∇, se aplica a funciones escalares (ψ(x, y, z)) y el resultado es un vector.<br>   
<br>
$$∇ ≡\frac{∂}{∂x}\hat{i}+\frac{∂}{∂y}\hat{j}+\frac{∂}{∂z}\hat{k}$$
<br>
$$ grad(ψ(x, y, z)) = ∇ψ(x, y, z) = \frac{∂ψ(x, y, z)}{dx}\hat{i}+\frac{∂ψ(x, y, z)}{dy}\hat{j}+\frac{∂ψ(x, y, z)}{dz}\hat{k}$$
